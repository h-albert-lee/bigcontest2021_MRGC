## 회사(업종) 데이터 meanshift로 밀도가 가장 높은 지점을 찾기

```python
result=[]   #업종데이터에 meanshift를 적용하기 위한 데이터로 변경
for i in range(len(comp_dong_x)):
    result.append([comp_dong_x[i],comp_dong_y[i]])
    
result
    
```

```python
from sklearn.cluster import estimate_bandwidth  #최적의 대역폭을 구하기

bandwidth = estimate_bandwidth(result)
print('bandwidth 값:', round(bandwidth,3))
best =round(bandwidth,3)
```

```python
import numpy as np
from sklearn.cluster import MeanShift

meanshift= MeanShift(bandwidth=0.007)
cluster_labels = meanshift.fit_predict(result)
print('cluster labels 유형:', np.unique(cluster_labels))
```

```python
cluster_labels
a=list(cluster_labels)
```

```python
from collections import Counter #Counter를 사용하여 각 군집별 얼마나 할당되었는지 확인 
result = Counter(a)
print(result)

result[0]
```

```python
cls=result.most_common(1)[0][0]  # 가장 할당이 많이 된 클러스터를 cls에 저장 
cls
```

```python
meanshift.cluster_centers_  #군집별 중심점 확인
```

```python
mil_x=meanshift.cluster_centers_[cls][0]  # cls의 위도 저장
mil_y=meanshift.cluster_centers_[cls][1] #cls의 경도 저장

print(mil_x,mil_y)
```

```python
only_cmp = all[all['type']==1]  #company(업종인거)
only_cmp
```

```python
import matplotlib.pyplot as plt    #업종데이터만 시각화하기 (meanshift를 활용하여 업종위치의 밀도중심 선정한거 시각화)
%matplotlib inline

only_cmp['meanshift_label']  = cluster_labels
centers = meanshift.cluster_centers_
unique_labels = np.unique(cluster_labels)
markers=['o', 's', '^', 'x', '*']


label_cluster = only_cmp[only_cmp['meanshift_label']==cls]
center_x_y = centers[cls]
# 군집별로 다른 마커로 산점도 적용
plt.scatter(x=only_cmp['위도'], y=only_cmp['경도'], edgecolor='k', marker=markers[cls] )

# 군집별 중심 표현
plt.scatter(x=center_x_y[0], y=center_x_y[1], s=200, color='gray', alpha=0.9, marker=markers[cls])
plt.scatter(x=center_x_y[0], y=center_x_y[1], s=70, color='k', edgecolor='k', marker='$%d$' % cls)

plt.show()
```

```python
radius = 0.000000001       #업종이 밀집된 밀도중심에서 반경을 넓혀가면서 빌딩을 선택 
flag = 0
result_building_location = {"x":[],"y":[], "i":[], "dist":[]}

while True:
    for i in range(0, len(building_dong_x)):
        distance = np.sqrt((building_dong_x[i] - mil_x) ** 2 + (building_dong_y[i] - mil_y) ** 2)
        #print('dis', distance)
        #print('radi', radius)
        if distance <= radius:
            if i in result_building_location['i']:
                continue
            print('ok')
            print(i,'번째', building_dong_x[i], building_dong_y[i])
            result_building_location['x'].append(building_dong_x[i])
            result_building_location['y'].append(building_dong_y[i])
            result_building_location['i'].append(i)
            result_building_location['dist'].append(distance)
            flag += 1
    if flag >= 1:
        break
    else:
        radius += 0.0002
        print('현재 radius', radius)
```

```python
result_building_location  #선정된 빌딩의 위도,경도,밀도중심에서의 거리
```

```python
here_building =here_building.drop(['연면적','주차장','복합주용도','평당금액2'],axis=1)
here_building
```

편의성 지수를 계산하기위해 지하철거리, 종합병원거리, 주변 대규모 점포 수, dist 컬럼에 standartscaler적용

```python
from sklearn.preprocessing import StandardScaler

# 객체생성
scaler = StandardScaler()

A_n = scaler.fit_transform(here_building['지하철거리(m)'].values.reshape(-1,1))
here_building.insert(0, '지하철거리_scaled', A_n)

scaler = StandardScaler()
A_n = scaler.fit_transform(here_building['종합병원거리(m)'].values.reshape(-1,1))
here_building.insert(0, '종합병원거리_scaled', A_n)

scaler = StandardScaler()
A_n = scaler.fit_transform(here_building['대로변거리(m)'].values.reshape(-1,1))
here_building.insert(0, '대로변거리_scaled', A_n)

scaler = StandardScaler()
A_n = scaler.fit_transform(here_building['주변 대규모 점포 수'].values.reshape(-1,1))
here_building.insert(0, '대규모점포수_scaled', A_n)

scaler = StandardScaler()
A_n = scaler.fit_transform(here_building['dist'].values.reshape(-1,1))
here_building.insert(0, 'dist_scaled', A_n)
# df.drop(['A'], axis=1, inplace=True)
```

```python
here_building
```

```python
here_building['편의성']= here_building['대로변거리_scaled']+here_building['종합병원거리_scaled']+here_building['지하철거리_scaled']-here_building['대규모점포수_scaled']
here_building  #편의성 지표생성
```

```python
here_building['totals']=1.8*here_building['편의성']+1.5*here_building['dist_scaled']
here_building # 편의성지표와 밀도중심에서의 거리를 모두 고려한 totals 컬럼생성  (작을수록 좋은거)
```

```python
here_building = here_building.sort_values('totals')  # totals를 기준으로 sort/ 최종 건물추천
```
