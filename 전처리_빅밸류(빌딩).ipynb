# 빅밸류의 산단데이터와 서울시 건물의 위경도 데이터를 MERGE하기 위한 전처리

```python
## 데이터 : 빅밸류/IND_REGINFO_202105.csV

import pandas as pd
industry  = pd.read_csv('/home/jovyan/WIDE_HOME/CompetitionData/빅밸류/IND_REGINFO_202105.csv',sep='|',encoding='cp949')
industry.head()
```

```python
industry=industry[['REG_ID','IC_ADMIN','IND_SIZE','IND_TYPE','ADDR']]  # 필요한 열만 선택
industry
```

```python
industry=industry.dropna()
seoul_ind = industry[industry['IC_ADMIN'].str.contains('서울')]
seoul_ind  #서울데이터만 추출
```

```python
# 데이터 : 서울특별시에 있는 건물에 대한 전체 위경도 데이터 (add라고 지정)
add=pd.read_excel('/home/jovyan/WIDE_HOME//유수빈/add.xlsx')
add
```

```python
def cleansing(text):     # seoul_ind의 ADDR컬럼의 괄호와 괄호안의 글자 제거를 위한 함수 지정
    pattern = '\(.*\)|\s-\s.*' 
    text = re.sub(pattern=pattern, repl='', string=text)
    if ',' in text:
        i=text.rfind(',')
        return text[:i]
    else:
        return text
```

```python
import re
for i in range(len(seoul_ind)):
    seoul_ind[i:i+1]['ADDR']=cleansing(seoul_ind.iloc[i]['ADDR']).rstrip()
    
seoul_ind
```

```python
def cleansing3(text):   # ADDR의 서울특별시 ~~구를 지우고 도로명만 남기기위한 함수 지정
    i = text.rfind('구 ')
    return text[i+1:]


for i in range(len(seoul_ind)):
    seoul_ind[i:i+1]['ADDR']=cleansing3(seoul_ind.iloc[i]['ADDR']).rstrip()
    
seoul_ind
```

```python
def cleansing2(text):  # 서울시 건물의 위경도 데이터에서 의미없는 0값이 들어있어 제거하기 위한 함수
    i = text.rfind('0')
    return text[:i-1]


for i in range(len(add)):
    add[i:i+1]['주소']=cleansing2(add.iloc[i]['주소']).rstrip()
    
add
```

```python
def cleansing4(text):  # 서울시 건물의 위경도 데이터에서 시,구,동을 제거하기 위한 함수
    i = text.rfind('동 ')
    return text[i+1:]


for i in range(len(add)):
    add[i:i+1]['주소']=cleansing4(add.iloc[i]['주소']).rstrip()
    
add
```

# 데이터 MERGE

```python
merge = pd.merge(seoul_ind,add, how='left', left_on='ADDR', right_on='주소')
```

```python
merge_not_null = merge.dropna()
merge_not_null
```

```python
import re  #기존 사업체 창업률 데이터에서도 동의 숫자제거

def remove_num(text):
    a= re.sub(r'[0-9]+', '', text)
    return a


for i in range(len(attr)):
    attr[i:i+1]['동']=remove_num(attr.iloc[i]['동']).rstrip()
    
attr
```

```python
import numpy as np  #사업체 창업률을 sub에 각 해당하는 동에 맞게 매칭
a=[]
ans = []
for i in sub[0]:
    for j in range(len(attr)):
        if attr.iloc[j]['동'] in i:
            a.append(int(attr.iloc[j]['사업체 창업률']))
    ans.append(round(np.mean(a),2))
    
    
ans
    
```
