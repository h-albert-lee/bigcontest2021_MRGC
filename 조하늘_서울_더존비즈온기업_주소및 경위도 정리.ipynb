# 더존비즈온


```python
# 회사 정보
comp  = pd.read_csv('./CompetitionData/더존비즈온/tbl_company.csv',sep='|') # 119621 rows × 7 columns
comp.columns = ['회사정보 가명화 연계키', '주소 시도', '주소 시군구', '주소 동', '표준 산업분류코드 대분류',
       '표준 산업분류코드 중분류', '법인 개인 여부']
comp # 119621 rows × 7 columns
```

```python
def industry(x):
  if ('도매 및 소매업' in str(x)):
    return '도매 및 소매업'
  elif ('사업시설 관리, 사업 지원 및 임대 서비스업' in str(x)):
    return '사업시설 관리, 사업 지원 및 임대 서비스업'
  elif ('부동산업' in str(x)):
    return '부동산업'
  elif ('전문, 과학 및 기술 서비스업' in str(x)):
    return '전문, 과학 및 기술 서비스업'
  elif ('금융 및 보험업' in str(x)):
    return '금융 및 보험업'
  elif ( '교육 서비스업' in str(x)):
    return  '교육 서비스업'
  elif ('정보통신업' in str(x)):
    return '정보통신업' 
  else:
    return '구분x'

comp['표준 산업분류코드 대분류']= comp['표준 산업분류코드 대분류'].apply(industry) 
comp.head()
```

```python
comp= comp[comp['표준 산업분류코드 대분류']!='구분x']
comp
```

```python
def sido(x):
  if ('서울특별시' in str(x)):
    return '서울특별시' 
  else:
    return '구분x'

comp['주소 시도']= comp['주소 시도'].apply(sido) 
```

```python
comp = comp[comp['주소 시도']!='구분x']
comp # 16438 rows × 7 columns
```
